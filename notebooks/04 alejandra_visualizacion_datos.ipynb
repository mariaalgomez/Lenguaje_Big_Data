```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Scatter plot (diagrama de dispersion)
x= np.arange(1,20)
y= x**2
y1=x**2+5*x

plt.scatter(x, y, s=40,color='violet', marker='*', label ='x^2')
plt.scatter(x, y1, s=40,color='blue', marker='o', label ='x**2+5*x')
plt.legend(title = 'Funciones')
plt.title('Tipos de funciones',fontsize =20,color='violet')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
#plt.xlim(0,20)
plt.text(5,300,"este texto",color='red')

#agregar las coordenadas de los marcadores en texto
for val_x in x:
    plt.text(val_x, val_x**2,f'({val_x},{val_x**2})')
```

```python
#histogramas
media =20
std =3
numeros =np.random.normal(media, std, 1000)

plt.hist(numeros, bins=20,cumulative=False,density=False)
plt.title('Distribucion normal')
plt.xlabel('numeros')
plt.ylabel('frecuencia')
```
